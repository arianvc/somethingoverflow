```python
import networkx as nx
from random import choice, random
import matplotlib.pyplot as plt
%matplotlib inline
import pygraphviz
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
# Make a random directed graph
G = nx.DiGraph()
nodes = range(10,21)
size = [random()*1000 for i in nodes]*2
G.add_nodes_from(nodes)
for i in range(10):
    G.add_edge(choice(nodes), choice(nodes), weight=int(10*2**random()))
pos = graphviz_layout(G, prog="neato")
colors = 'rgby'
nx.draw(G, pos, node_size=size, node_color=colors, with_labels=True)
#TODO pictures

# Saving the graph
plt.savefig('random.svg')
```
